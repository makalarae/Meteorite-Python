```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('meteorites.csv')
df2=pd.read_csv('11.csv')
df3=pd.read_csv('12.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
df1_and_df3 = pd.merge(df1, df3[['name', 'classification_id']], on='name', how='left')
# joins on name
```

```python
# After joining on name join again on classification_id
df_final = pd.merge(df1_and_df3, df2[['classification_id', 'category']], on='classification_id', how='left')
```

```python
df_final.head()
```

```python
# replaces null with 0
df_final['classification_id'] = df_final['classification_id'].fillna(0).astype(int)
# convert to int with null values for 0
df_final['classification_id'] = df_final['classification_id'].astype('Int64')
```

```python
#need to drop the extra \r 
df_final['category'] = df_final['category'].str.replace('\r', '')
```

```python
# export to meteorite.csv for 1 dataframe
df_final.to_csv('meteorite.csv', index=False)
```
